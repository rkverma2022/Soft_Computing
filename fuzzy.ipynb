```python
class membershipFunction():
    def A(x):
        if x < 30:
            return 1
        elif x > 30 and x < 50 :
            return (x - 30)/(50 - 30)
        elif x > 50 and x < 70 :
            return (x - 50)/ (70 - 50)
        else:
            return 0
    
    def B(x):
        if x < 60 :
            return 0
        elif x > 60 and x < 80 : 
            return (x - 60)/(80 - 60)
        elif x > 80 and x < 100 :
            return (x - 80)/(100 - 80)
        else:
            return 1

            
```

```python
universeOfDiscourse = []
for i in range(1, 16):
    universeOfDiscourse.append(i * 10)
universeOfDiscourse
    
```

```python
#Ya and Yb, just for my reference
Ya = []
Yb = []


Ua = set()
Ub = set()
for x in universeOfDiscourse:
    f  = membershipFunction 
    a = f.A(x=x)
    b = f.B(x=x)
    
    Ua.add(a)
    Ub.add(b)
    
    Ya.append(a)
    Yb.append(b)
Ub
    
```

```python
Ya, Yb
```

```python
plt.title("Membership Curve of fuzzy set : A(x) and B(x)")
plt.plot(universeOfDiscourse, Ya)

plt.plot(universeOfDiscourse, Yb)
plt.xlabel("universe of discourse - x")
plt.ylabel("membership value for A and B")

plt.legend([("A(x) : ", Ya), (("B(x) :  ", Yb))])
```

<p style="color: red; item-align: centre; text-align: centre; padding-left: 30%"> 1. C(x) = A(x) U B(x) </p>

```python
Yc = []

for i in range(len(Ya)):
    if Ya[i] > Yb[i]:
        Yc.append(Ya[i])
    else:
        Yc.append(Yb[i])
```

```python
plt.plot(universeOfDiscourse, Yc)
plt.title("C(x) = A(x) U B(x)")
plt.ylabel("Membership Value of fuzzy set C(x)")
plt.xlabel("universe of discourse - x")
plt.legend([Yc])
```

<p style="color: red; item-align: centre; text-align: centre; padding-left: 30%; font-family: console"> 2. D(x) = A(x) intersection B(x) </p>

```python
# Dx = Ua.intersection(Ub)

Yd = []

for i in range(len(Ya)):
    if Ya[i] < Yb[i]:
        Yd.append(Ya[i])
    else:
        Yd.append(Yb[i])
```

```python
plt.plot(universeOfDiscourse, Yd)
plt.title("D(x) = A(x) intersection B(x)")
plt.xlabel("universe of discourse")
plt.ylabel("membership value")
plt.legend([Yd])
```

<h3 style="color: green; padding-left: 40% "> Question 2</h3>

```python
class temperatureMemebershipFunction:
    def C(x):
        if x < -10:
            return 0
        elif x > -10 and x < 0:
            return (x - (-10))/(0-(-10))
        elif x >= 0 and x < 5:
            return 1
        else:
            return 0
    
    def D(x):
        if x < 25:
            return 0
        elif x < 35:
            return (x - 25)/(35 - 25)
        elif x < 40:
            return 1
        else:
            return 0
        
        
```

```python
universeOfDiscourseTemperature = []
for i in range(-2, 11):
    
    universeOfDiscourseTemperature.append(i * 5)
universeOfDiscourseTemperature
    
```

```python

#for reference
Rc = []
Rd = []

Uc = set()
Ud = set()

for x in universeOfDiscourseTemperature:
    
    g = temperatureMemebershipFunction
    
    c = g.C(x)
    d = g.D(x)
    
    Rc.append(c)
    Rd.append(d)
    
    Uc.add(c)
    Ud.add(d)
    
    
    
```

```python
MaxMinComp = []

for i in range(len(Rc)):
    MaxMinComp.append(max(min(Rc[i], Rd[i]), Rc[i]))
```

```python
plt.title("Rc | Rd Membership Graph")
plt.plot(universeOfDiscourseTemperature, Rc)
plt.plot(universeOfDiscourseTemperature, Rd)

plt.legend([("C(x) ",Rc),("D(x) ", Rd)])
plt.xlabel("universe of discourse")
plt.ylabel("membership value")
```

```python
plt.title("Max Min Composition ")
plt.plot(universeOfDiscourseTemperature, MaxMinComp)
plt.xlabel("universeOfDiscourseTemperature")
plt.ylabel("Min Max Composition")
plt.legend([MaxMinComp])
```
