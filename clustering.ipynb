```python
data = df.to_numpy()
dissimilarity_matrix = np.zeros((data.shape[0], data.shape[0]))
dissimilarity_matrix.shape
```

```python
def euclidian_distance(pt1, pt2):
    d = pt1-pt2
    X = 0
    for a in d:
        X += a**2
    return np.math.sqrt(X)
```

```python
# pt1 = np.array([0,0, 0, 0])
# pt2 = np.array([10, 10, 10,10])

# euclidian_distance(pt1, pt2)
```

```python

for i in range(data.shape[0]):
    for j in range(i, data.shape[0]):
        d = euclidian_distance(data[i], data[j])
        dissimilarity_matrix[i, j] = d
        dissimilarity_matrix[j, i] = d
```

```python
eucli_dissimilarity = pd.DataFrame(dissimilarity_matrix)
eucli_dissimilarity.to_excel("Euclidean_Dissimilarity.xlsx")
eucli_dissimilarity
```

```python
def remove_subsets(new_Clusters):
    CLUSTER = new_Clusters.copy()
    
    for i in range(len(new_Clusters)):
        s1 = new_Clusters[i]
        for j in range(i + 1, len(new_Clusters)):
            s2 = new_Clusters[j]
            try:
                if s1.issubset(s2):
                    CLUSTER.remove(s2)
                elif s2.issubset(s1):
                    CLUSTER.remove(s1)
            except:
                continue
    
    return CLUSTER
```

```python
def similarity_of_P_Clusters(CLUSTER):
    p = len(CLUSTER)
    Similarity_of_P_C = np.zeros((p, p))
    for i in range(p):
        s1 = CLUSTER[i]
        for j in range(i, p):
            s2 = CLUSTER[j]
            value = len(s1.intersection(s2))/len(s1.union(s2))
            Similarity_of_P_C[i, j] = value
            Similarity_of_P_C[j, i] = value
    return Similarity_of_P_C
```

```python
def pairing_of_most_similar_clusters(Similarity_of_P_C):
    pair_of_set = list()
    print("pairs of most similar clusters")
    
    p = Similarity_of_P_C.shape[0]
    for i in range(p):
        max = -100
        k = 0
        l = 0
        for j in range(i , p):
            if Similarity_of_P_C[i, j] == 1:
                continue
            elif max < Similarity_of_P_C[i, j]:
                max = Similarity_of_P_C[i, j]
                k = i
                l = j
        if(k!=l):
            pair_of_set.append((k, l))

    Ckl = []
    for sets in pair_of_set:
        s1 = new_Clusters[sets[0]]
        s2 = new_Clusters[sets[1]]

        s = s1.union(s2)
        Ckl.append(s)
    
    new_clusters = Ckl
    return new_clusters

```
