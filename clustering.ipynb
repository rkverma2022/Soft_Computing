<div style="font-family: console; color: green; padding-left: 4%; padding-right: 5%"> <p>2. (A). Create a similarity matrix S of size m×m <b style= "color: red">(150, 150) </b> where, each (i, j)-th entry in the matrix gives the
dissimilarity measurement between i-th and j-th objects. Use Eucledian distance to measure
the dissimilarity</p> </div>

```python
data = df.to_numpy()
dissimilarity_matrix = np.zeros((data.shape[0], data.shape[0]))
dissimilarity_matrix.shape
```

```python
def euclidian_distance(pt1, pt2):
    d = pt1-pt2
    X = 0
    for a in d:
        X += a**2
    return np.math.sqrt(X)
```

```python
# pt1 = np.array([0,0, 0, 0])
# pt2 = np.array([10, 10, 10,10])

# euclidian_distance(pt1, pt2)
```

```python

for i in range(data.shape[0]):
    for j in range(i, data.shape[0]):
        d = euclidian_distance(data[i], data[j])
        dissimilarity_matrix[i, j] = d
        dissimilarity_matrix[j, i] = d
```

```python
eucli_dissimilarity = pd.DataFrame(dissimilarity_matrix)
eucli_dissimilarity.to_excel("Euclidean_Dissimilarity.xlsx")
eucli_dissimilarity
```
