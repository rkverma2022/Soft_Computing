```python
class membershipFunction():
    def A(x):
        if x < 30:
            return 1
        elif x > 30 and x < 50 :
            return (x - 30)/(50 - 30)
        elif x > 50 and x < 70 :
            return (x - 50)/ (70 - 50)
        else:
            return 0
    
    def B(x):
        if x < 60 :
            return 0
        elif x > 60 and x < 80 : 
            return (x - 60)/(80 - 60)
        elif x > 80 and x < 100 :
            return (x - 80)/(100 - 80)
        else:
            return 1

            
```

```python
universeOfDiscourse = []
for i in range(1, 16):
    universeOfDiscourse.append(i * 10)
universeOfDiscourse
    
```

```python
#Ya and Yb, just for my reference
Ya = []
Yb = []


Ua = set()
Ub = set()
for x in universeOfDiscourse:
    f  = membershipFunction 
    a = f.A(x=x)
    b = f.B(x=x)
    
    Ua.add(a)
    Ub.add(b)
    
    Ya.append(a)
    Yb.append(b)
Ub
    
```

<p style="color: red; item-align: centre; text-align: centre; padding-left: 30%"> 1. C(x) = A(x) U B(x) </p>

```python
Cx = Ua.union(Ub)
Cx
```

```python
c = list(Cx)

c
```

```python
plt.plot(c)
plt.title("Membership Curve of fuzzy set : C(x) = A(x) U B(x)")
```

```python
plt.title("Membership Curve of fuzzy set : A(x) and B(x)")
plt.scatter(Ya, universeOfDiscourse)
plt.legend("A")
plt.scatter(Yb, universeOfDiscourse)
plt.xlabel("universe of discourse - x")
plt.ylabel("membership value for A and B")
# plt.legend("B")
```

<p style="color: red; item-align: centre; text-align: centre; padding-left: 30%; font-family: console"> 2. D(x) = A(x) intersection B(x) </p>

```python
Dx = Ua.intersection(Ub)
```

```python
print(Dx)
plt.plot(list(Dx))
plt.title("Membership Curve of fuzzy set : Dx = A(x) intersection B(x)")

```

<h3 style="color: green; padding-left: 40% "> Question 2</h3>

```python
class temperatureMemebershipFunction:
    def C(x):
        if x < -10:
            return 0
        elif x > -10 and x < 0:
            return (x - (-10))/(0-(-10))
        elif x >= 0 and x < 5:
            return 1
        else:
            return 0
    
    def D(x):
        if x < 25:
            return 0
        elif x < 35:
            return (x - 25)/(35 - 25)
        elif x < 40:
            return 1
        else:
            return 0
        
        
```

```python
universeOfDiscourseTemperature = []
for i in range(-2, 11):
    
    universeOfDiscourseTemperature.append(i * 5)
universeOfDiscourseTemperature
    
```

```python

#for reference
Rc = []
Rd = []

Uc = set()
Ud = set()

for x in universeOfDiscourseTemperature:
    
    g = temperatureMemebershipFunction
    
    c = g.C(x)
    d = g.D(x)
    
    Rc.append(c)
    Rd.append(d)
    
    Uc.add(c)
    Ud.add(d)
    
    
    
```

```python
Mc = np.matrix(list(Uc), dtype=float)
Md = np.matrix(list(Ud), dtype=float).transpose()

Mc
```

```python
Md
```

```python
M = np.dot(Md, Mc)
```

```python
M
```

```python
max_min_composition = []

for i in range(M.shape[0]):
    m = 100
    for j in range(M.shape[1]):
        m = min(m, M[i,j])
    m = max(m, list(Uc)[i])
    max_min_composition.append(m)
```

```python
Fx = max_min_composition
Fx
```

```python
plt.plot(Fx)
plt.title("membership curve of F(x) = max (min(C(x), D(x)))")
```
